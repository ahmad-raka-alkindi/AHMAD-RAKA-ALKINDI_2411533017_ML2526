```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 2. Generate data dummy
data = {'luas': [6, 8, 10, 14, 18], 'harga': [7, 9, 13, 17.5, 18]}
df = pd.DataFrame(data)
print(df)

# 3. Visualisasi scatterplot
df.plot(kind='scatter', x='luas', y='harga')
plt.title('Perbandingan Luas Tanah dengan Harga Tanah')
plt.xlabel('Luas Tanah (m persegi)')
plt.ylabel('Harga Tanah (dollar)')
plt.grid(True)
plt.show()

# 4 & 5. Pisahkan variabel dan Latih Model
X = np.array(df['luas']).reshape(-1, 1)
y = np.array(df['harga'])
model = LinearRegression()
model.fit(X, y)
print(f'intercept: {model.intercept_}')
print(f'slope: {model.coef_}')

# 6. Visualisasi garis persamaan linear
X_vis = np.array([0, 25]).reshape(-1, 1)
y_vis = model.predict(X_vis)
plt.scatter(X, y)
plt.plot(X_vis, y_vis, color='red')
plt.title('Perbandingan Luas Tanah dengan Harga Tanah')
plt.grid(True)
plt.show()

# 7 & 8. Prediksi harga rumah
luas_tanah = np.array([12, 15, 20]).reshape(-1, 1)
harga_prediksi = model.predict(luas_tanah)
for luas, hrg in zip(luas_tanah, harga_prediksi):
    print(f'Luas Tanah {luas}: harga prediksi: {hrg}')
    from google.colab import drive
from google.colab import drive

```
